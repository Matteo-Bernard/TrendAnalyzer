```python
import trendist
import yfinance as yf
```

```python
data = yf.download('^GSPC', start='2000-01-01', end='2025-01-01')['Close']['^GSPC']
trend_analyzer = trendist.Trends(data, threshold=0.1, timeperiod='D')
trend_analyzer.fit()
```

```python
trend_analyzer.plot_trend(start='2023-01-01', end='2025-01-01', logy=False)
```

```python
trend_analyzer.get_proba()
```

```python
trend_analyzer.plot_distrib()
```

```python
trend_analyzer.plot_cumdistrib()
```

```python
import pandas as pd
from trendist_toolkit import trends
import yfinance as yf
%load_ext autoreload
%autoreload 2
```

```python
market_df = yf.Ticker('^FCHI').history('max').Close
market_df.index = pd.to_datetime(market_df.index.date)
market_df.name = '^FCHI'
```

```python
threshold = 0.1
trend = trends(market_df, threshold, timeperiod='W').fit()
```

```python
metric = 'Length'
trend.plot_trend()
trend.get_proba(metric)
trend.plot_distrib(metric)
trend.plot_cumdistrib(metric)
```

```python
import pandas as pd
from trendist_toolkit import trends 
import yfinance as yf
%load_ext autoreload
%autoreload 2
```

```python
history = yf.Ticker('^FCHI').history('10y').Close
history.index = pd.to_datetime(history.index, utc=True)
history.index = history.index.tz_localize(None)
```

```python
trend = trends(history=history, threshold=0.05, timeperiod='W').fit()
```

```python
trend.plot_trend()
```

```python
trend.get_proba('Length')
```

```python
trend.plot_distrib('Length')
```

```python
trend.plot_cumdistrib('Length')
```
